```sas
*Ex5_how_many_obs.sas;
*Method 1;
options nonotes nosource nonumber nodate leftmargin=1cm ;
DATA  _NULL_;
 SET sashelp.heart NOBS=numobs;
 if numobs then PUT '---> ' @7 "Number of cases =" numobs comma7.;
 stop;
run;
```

```sas
*Method 2;
options nonotes nosource  nonumber nodate leftmargin=1cm;
DATA   _NULL_;
  SET sashelp.heart END=last;
  count+1;
  if last then PUT '---> ' @7 "Number of cases =" count comma7.;
run;
```

```sas
*Method 3;
options nonotes nosource nonumber nodate leftmargin=1cm ;
%let s=----->;
DATA  _NULL_;
 if 0 then SET sashelp.heart NOBS=N;
   CALL SYMPUTX('total', N);
 stop;
run;
/* Below are 3 ways to display the value of the macro variable (&total) */
%PUT &s &total;
%PUT &s Number of cases = %SYSFUNC(left(&total));
%PUT &s Number of cases = %SYSFUNC(left(%qsysfunc(putn(&total, comma7.))));
```

```sas
*Method 4;
options nonotes nosource nonumber nodate leftmargin=1cm ;
PROC SQL noprint;
select count(*)into :OBSCOUNT
 from sashelp.heart;
 quit;
%PUT &s Number of cases = %SYSFUNC(left(%qsysfunc(putn(&total, comma7.))));
```
