### Copy .SAS files from a particular folder to a new folder

```python
import glob, os, shutil

files = glob.iglob(os.path.join('U:\A_DataRequest', "*.sas"))
for file in files:
    if os.path.isfile(file):
        shutil.copy2(file, 'U:\All_SAS')
```

### Move .SAS files from a particular folder to a new folder

```python
import os
import shutil
sourcepath='U:/_Misc/'
sourcefiles = os.listdir(sourcepath)
destinationpath = 'U:/All_SAS'
for file in sourcefiles:
    if file.endswith('.sas'):
        shutil.move(os.path.join(sourcepath,file), os.path.join(destinationpath,file))
```

### List out the name of the file if it exist in a certain folder

```python
import glob
path = 'C:\\test'
files = (f for f in glob.glob(path + '**/autoexec.sas', recursive=True))
for f in files:
    print(f)
```

### List out the file that exists by the same name in different folders

```python
import os
path = 'C:\\'
files = []
for r, d, f in os.walk(path):
    for file in f:
        if 'autoexec.sas' in file:
            files.append(os.path.join(r, file))
            
for f in files:
    print(f)
```
