##### Convert a SAS data set into a pandas dataframe

```python
import saspy
sas=saspy.SASsession()
cars= sas.sd2df('cars', libref='sashelp')
```

```python
type(cars)
```

```python
import pandas as pd
df_x = pd.DataFrame(cars,columns=['Make','Type','Invoice'])
df_x.head()
```
