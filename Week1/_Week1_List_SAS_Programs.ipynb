
#### How to list files with a common extension from a specified directory but no directories or subdirectories? (Method 1)

```python
import pandas as pd
from pathlib import Path
import time

p = Path("C:/SASCourse/Week1")
all_files = []
for i in p.rglob('*.SAS'):
    all_files.append((i.name, time.ctime(i.stat().st_ctime)))

columns = ["File_Name", "Created"]
df = pd.DataFrame.from_records(all_files, columns=columns)
xdf=df.iloc[:,[0]]
print(xdf.to_string(index=False))
```

#### How to list files with a common extension from a specified directory but no directories or subdirectories? (Method 2)

```python
import glob
path = 'C:\\SASCourse\Week1\\'
files = (f for f in glob.glob(path + '**/*.sas', recursive=True))
for f in files:
    print(f)
```
